```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv("basic_features.csv")
df.head()
```

```python
df.columns
```

```python
features = [
    "length",
    "GC_like",
    "identity_to_human_y"
]

# Load the additional features from 'tp53_features_with_similarity.csv'
df_additional_features = pd.read_csv("tp53_features_with_similarity.csv")

# Merge the additional features into the main dataframe, explicitly handling suffixes
df = pd.merge(df, df_additional_features, on="id", how="left", suffixes=('', '_y'))

# Debugging: Print columns of both dataframes after merge
print("Columns in df_additional_features:", df_additional_features.columns)
print("Columns in df after merge:", df.columns)

scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df[features])

# Simple, explainable score
df["importance_score"] = df_scaled.mean(axis=1)

# Rank codons
df_ranked = df.sort_values("importance_score", ascending=False)

df_ranked.head()
```

```python
df_ranked[["id", "importance_score"]].to_json(
    "ml_hotspot_ranking.json",
    orient="records",
    indent=2
)
```
